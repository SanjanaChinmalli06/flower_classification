```python
def predict_iris_species(sepal_length, sepal_width, petal_length, petal_width, model='lr'):
    """
    Predict Iris species for given measurements.
    
    Parameters:
    -----------
    sepal_length : float
        Length of the sepal in cm
    sepal_width : float
        Width of the sepal in cm
    petal_length : float
        Length of the petal in cm
    petal_width : float
        Width of the petal in cm
    model : str
        Model to use for prediction ('lr' for Logistic Regression, 'dt' for Decision Tree)
    
    Returns:
    --------
    str : Predicted species name
    """
    
    # Create input array
    input_features = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    
    # Select model
    selected_model = lr_model if model == 'lr' else dt_model
    
    # Make prediction
    prediction = selected_model.predict(input_features)[0]
    
    # Map prediction back to species name
    species_map = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
    
    return species_map[prediction]

# Test the prediction function with example inputs
print("="*60)
print("TESTING PREDICTION FUNCTION")
print("="*60)

# Example 1: Typical Setosa
pred1_lr = predict_iris_species(5.1, 3.5, 1.4, 0.2, model='lr')
pred1_dt = predict_iris_species(5.1, 3.5, 1.4, 0.2, model='dt')
print(f"\nExample 1 - Typical Setosa measurements (5.1, 3.5, 1.4, 0.2):")
print(f"  Logistic Regression: {pred1_lr}")
print(f"  Decision Tree: {pred1_dt}")

# Example 2: Typical Versicolor
pred2_lr = predict_iris_species(5.9, 3.0, 4.2, 1.5, model='lr')
pred2_dt = predict_iris_species(5.9, 3.0, 4.2, 1.5, model='dt')
print(f"\nExample 2 - Typical Versicolor measurements (5.9, 3.0, 4.2, 1.5):")
print(f"  Logistic Regression: {pred2_lr}")
print(f"  Decision Tree: {pred2_dt}")

# Example 3: Typical Virginica
pred3_lr = predict_iris_species(6.3, 3.3, 6.0, 2.5, model='lr')
pred3_dt = predict_iris_species(6.3, 3.3, 6.0, 2.5, model='dt')
print(f"\nExample 3 - Typical Virginica measurements (6.3, 3.3, 6.0, 2.5):")
print(f"  Logistic Regression: {pred3_lr}")
print(f"  Decision Tree: {pred3_dt}")

print("\n" + "="*60)
```

## 8. Build Prediction Script / CLI for New Input

```python
# Detailed classification reports
print("\n" + "="*60)
print("LOGISTIC REGRESSION - DETAILED CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_test, y_pred_lr, 
                          target_names=['Setosa', 'Versicolor', 'Virginica']))

print("\n" + "="*60)
print("DECISION TREE - DETAILED CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_test, y_pred_dt,
                          target_names=['Setosa', 'Versicolor', 'Virginica']))

# Analyze misclassifications
print("\n" + "="*60)
print("MISCLASSIFICATION ANALYSIS")
print("="*60)

lr_misclassified = y_test != y_pred_lr
dt_misclassified = y_test != y_pred_dt

print(f"\nLogistic Regression:")
print(f"  Total misclassifications: {lr_misclassified.sum()}")
if lr_misclassified.sum() > 0:
    print(f"  Misclassification rate: {lr_misclassified.sum() / len(y_test) * 100:.2f}%")
    misclass_indices = np.where(lr_misclassified)[0]
    print(f"  Sample misclassified instances: {misclass_indices[:5]}")

print(f"\nDecision Tree:")
print(f"  Total misclassifications: {dt_misclassified.sum()}")
if dt_misclassified.sum() > 0:
    print(f"  Misclassification rate: {dt_misclassified.sum() / len(y_test) * 100:.2f}%")
    misclass_indices = np.where(dt_misclassified)[0]
    print(f"  Sample misclassified instances: {misclass_indices[:5]}")

print("\n" + "="*60)
print("KEY INSIGHTS:")
print("="*60)
print("• Setosa is clearly separable from other species")
print("• Versicolor and Virginica show some overlap")
print("• Logistic Regression uses linear decision boundaries")
print("• Decision Tree can capture non-linear patterns")
print("• Most errors occur between Versicolor and Virginica")
print("="*60)
```

```python
# Compute confusion matrices
cm_lr = confusion_matrix(y_test, y_pred_lr)
cm_dt = confusion_matrix(y_test, y_pred_dt)

# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Logistic Regression Confusion Matrix
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0], 
            xticklabels=['Setosa', 'Versicolor', 'Virginica'],
            yticklabels=['Setosa', 'Versicolor', 'Virginica'],
            cbar_kws={'label': 'Count'})
axes[0].set_title('Logistic Regression - Confusion Matrix', fontsize=12, fontweight='bold')
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')

# Decision Tree Confusion Matrix
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', ax=axes[1],
            xticklabels=['Setosa', 'Versicolor', 'Virginica'],
            yticklabels=['Setosa', 'Versicolor', 'Virginica'],
            cbar_kws={'label': 'Count'})
axes[1].set_title('Decision Tree - Confusion Matrix', fontsize=12, fontweight='bold')
axes[1].set_ylabel('True Label')
axes[1].set_xlabel('Predicted Label')

plt.tight_layout()
plt.show()
```
